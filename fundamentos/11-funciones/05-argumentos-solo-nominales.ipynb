## Argumentos solo nominales
Existe el parametro `*` que sirve de limite en una función. 
Todos los parametros despues de `*` esta obligados a ser parametros nominales.



```python
def fn_parametros_nominales(nombre, apellido, *, edad=0):
    if edad >= 18:
        return dict(
            nombre = nombre,
            apellido = apellido
        )
    else:
        return dict(
            nombre = nombre
        )

print(fn_parametros_nominales("Cris", "Villca", edad=25))
print(fn_parametros_nominales("Cris", "Villca"))
print(fn_parametros_nominales("Cris", "Villca", 25)) ## error! edad debe ser nominal
```
