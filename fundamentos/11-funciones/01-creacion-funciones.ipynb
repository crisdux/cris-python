## Definir funciones
En python usamos la palabra reservada `def` para definir funciones seguido del nombre de la fucin y entre parentesis los parametros que recive esa funcin separados por comas, posteriormente `:` que equivale a abrir la funcin y con una identacin dentro de la funcin esta el cuerpo de la misma, el cuerpo puede o no tener una sentencia `return`.



```python
def saludar(nombre:str):
    print(f"Hola {nombre}")

saludar("Cris")
```

Las fucionespor lo general regresan un valor para que pueda ser usado en otra parte del programa, para ello usamos `return`


```python
def suma(num1, num2):
    return num1 + num2

if suma(5, 2) > 1: ## usamos la llamada d ela funcin en una condicin 
    print(f"{suma(5,2)} es mayor que 1")
else:
    print(f"{suma(5,2)} no es mayor que 1")
```

## Regresando multiples valores
Cuando regresamos valors multiples reparados por comas lo que en realidad regresamos es una tupla, veamos

```python
def get_nombre_completo(nombre, apellido):
    return nombre, apellido ## es una tupla

print(get_nombre_completo("Cris", "Villca")) ## podemos desmpaquetar la tuplas
```

```python
nombre, apellido = get_nombre_completo("Cris", "Villca")
print(nombre)
print(apellido) ## ahora tenemos los valores separados
```
