## Operaciones
Podemos hacer las operaciones básicas de la teoria de conjuntos matematica



```python
A = {1, 2}
B = {2, 3}

I = A.intersection(B) ## solo los elementos que esten presentes en ambos sets
print(I)

U = A.union(B) ## todos los elementos de A y B sin repetir
print(U)

D = A.difference(B) ## quitamos en A todos los elementos que se repiten en B
print(D)

DS = A.symmetric_difference(B) ## elementos que estan en A y B pero no en ambos
print(DS)
```
