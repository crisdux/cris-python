## Consecutivo
 El objetivo es encontrar el primer número no consecutivo dentro de una lista de valores numéricos enteros. Si todos los valores son consecutivos entonces el resultado debe ser None.




```python
def consecutivo(values):
    res = None
    if not values:
        return None
    values_zip = [ value for value in range(values[0], values[-1]+1)]
    for value, value_zip in zip(values, values_zip):
        if value != value_zip:
            res = value
            break
    return res

print(consecutivo([1, 2, 3, 4, 6, 7, 8]))
print(consecutivo([101, 102, 103]))
print(consecutivo([-5, -4, -3, 0, 3, 4, 5]))
print(consecutivo([1]))
print(consecutivo([]))
```
