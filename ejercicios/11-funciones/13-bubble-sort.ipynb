## Bubble sort
Ordene una lista de entrada utilizando el algoritmo de la burbuja.



```python
def bubble_sort(numeros):
    for i in range(len(numeros)):
        for j in range(len(numeros)-i-1):
            if(numeros[j] > numeros[j+1]):
                aux = numeros[j]
                numeros[j] = numeros[j+1]
                numeros[j+1] = aux
    return numeros


print(bubble_sort([4, 9, 1, 5]))
print(bubble_sort([2,1]))
print(bubble_sort([1]))
print(bubble_sort([]))

#  [4, 9, 1, 5] original

#  [4, 9, 1, 5]
#  [4, 1, 9, 5]
#  [4, 1, 5, 9]

#  [1, 4, 5, 9]
#  [1, 4, 5, 9]

# [1, 4, 5, 9]

```
