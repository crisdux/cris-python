## Consecutivos
Escriba una función que reciba una lista de elementos y un parámetro "opcional" para el formato de salida. El objetivo es encontrar la frecuencia de aparición de elementos consecutivos en la lista de entrada.

Si el parámetro de salida como cadena de texto está a False se deberá devolver una lista de tuplas, donde cada tupla contiene el elemento y el número de repeticiones (frecuencia). Si el parámetro de salida como cadena de texto está a True se deberá devolver una cadena de texto con elemento:frecuencia separados por comas.

* Se debe obligar a que el primer parámetro (lista de elementos) sea pasado como posicional.
* El segundo parámetro para el formato de salida como cadena de texto debe ser falso por defecto.




```python
def consecutive(items, /, formato=False):
    res_dict = {}
    res_list = []
    for item in items:
        if item not in res_dict:
            res_dict[item] = 1
        res_dict[item] = res_dict[item] + 1
    
    if not formato:
        for key, value in res_dict.items():
            res_list.append((key, value))
        return res_list

    for key, value in res_dict.items():
        res_list.append(f"{key}:{value}")
    return ",".join(res_list)


print(consecutive([1, 2, 2, 2, 4, 4, 4, 5, 5, 5, 5])) ## {1: 1, 2: 3, 4: 3, 5: 4}
print(consecutive([1, 2, 2, 2, 4, 4, 4, 5, 5, 5, 5], True))
```
