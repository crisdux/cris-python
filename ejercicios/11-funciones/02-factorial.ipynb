## Factorial
Escriba una función que calcule el factorial de un número n. El factorial de un número n se define como:



* 0! = 1
* El factorial de un numero negativo no existe



```python
def factorial(n):
    if n < 0:
        return None
    res = 1
    for i in range(1, n+1):
        res *= i
    return res

print(factorial(5))
print(factorial(0))
print(factorial(-1))
    
```
