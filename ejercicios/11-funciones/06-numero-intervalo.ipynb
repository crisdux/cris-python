## Intervalo numero
Escriba una función en Python que indique si un número está en un determinado intervalo.
* Los límites deben tenerse en cuenta como parte del intervalo.
* El número debe ser obligatoriamente un parámetro posicional.



```python
def intervalo_numero(numero:int, /, limite_inferior:int, limite_superior:int)->bool:
    return limite_inferior <= numero <= limite_superior

print(intervalo_numero(3,2,5))
print(intervalo_numero(10,10,10))
print(intervalo_numero(-1,-10,10))
print(intervalo_numero(5,0,1))
    
```
