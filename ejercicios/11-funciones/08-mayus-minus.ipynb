## Separando mayusculas de minusculas
 Dada una lista de palabras, escriba una función que devuelva dos listas, la primera con las palabras que vienen en minúsculas y la segunda con las palabras que vienen en mayúsculas.

Ignore aquellas palabras que mezclan mayúsculas y minúsculas.



```python
def mayus_minus(palabras: list[str])-> list[str]:
    return [palabra for palabra in palabras if palabra.isupper()], [palabra for palabra in palabras if palabra.islower()]

print(mayus_minus(['AZUL', 'blanco', 'NEGRO', 'amarillo']))
print(mayus_minus(['FEDORA', 'debian', 'OPENSUSE', 'ubuntu', 'MacOS']))
print(mayus_minus(['python', 'c', 'javascript']))
```
