## Pangrama
 Escriba una funci√≥n en Python que determine si una cadena de texto es un pangrama.

 

```python
def pangrama(cadena):
    lista_base = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    cadena_a_lista = list(cadena)
    filtrar_espacios = [i.lower() for i in cadena_a_lista if i != " "]
    eliminar_espacios = list(set(filtrar_espacios))
    lista_ordenada = sorted(eliminar_espacios)
    return lista_ordenada == lista_base


print(pangrama('The quick brown fox jumps over the lazy dog'))
print(pangrama('Sylvia wagt quick den Jux bei Pforzheim'))
print(pangrama('Portez ce whisky au vieux juge blond qui fume'))
print(pangrama('No soy un pangrama'))
```
