## Lista de pares
Escriba una función en Python que reciba una lista de valores enteros y devuelva otra lista sólo con aquellos valores pares.



```python
def lista_pares(numeros:list[int])->list[int]:
    return [i for i in numeros if i % 2 == 0]

print(lista_pares([1, 2, 3, 4, 5, 6, 7, 8, 9]))
print(lista_pares([8, 6, 1, 2, 3, 4]))
print(lista_pares([]))
print(lista_pares([1, 3, 5]))
print(lista_pares([2, 2, 2]))
```
