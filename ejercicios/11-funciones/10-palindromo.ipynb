## Palindromo
Escriba una función en Python que determine si una cadena de texto es un palíndromo.



```python
def palindromo(cadena):
    a_lista = list(cadena.lower())
    sin_espacios = [i for i in a_lista if i != ' ']
    lista_reversa = sin_espacios[::-1]
    return "".join(lista_reversa) == "".join(sin_espacios)

print(palindromo('ana lava lana'))
print(palindromo('Yo dono rosas oro no doy'))
print(palindromo('yo soy'))
print(palindromo('La ruta natural'))
print(palindromo('No soy un palíndromo'))
```
