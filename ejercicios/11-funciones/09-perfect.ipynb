## Perfecto
Escriba una función en Python que indique si un número es perfecto. Utilice una función auxiliar que calcule los divisores propios.



```python
def perfecto(numero:int):
    contador = []
    for i in range(1, numero):
        if numero % i == 0:
            contador.append(i)
    return sum(contador) == numero

print(perfecto(28))
print(perfecto(496))
print(perfecto(512))
print(perfecto(1024))
print(perfecto(8128))
```
