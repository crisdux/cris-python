## Multiplicacion de Jack
A Jack le gusta hacer una multiplicación muy particular: Dado un número n multiplica n por 5 elevado al número de dígitos de n.
Ejemplo:

```py
 n = 10 --> 10 * 5²
```



````python
def multiplicacion_jack(n):
    len_n_str = len(str(n))
    return f"{n*(5**int(len_n_str))}"

print(multiplicacion_jack(3))
print(multiplicacion_jack(10))
print(multiplicacion_jack(200))
print(multiplicacion_jack(0))
print(multiplicacion_jack(-3))
````
