## Intercambio de nombre
Escriba un programa en Python que acepte el nombre y los apellidos de una persona y los devuelva en orden inverso separados por una coma. Utilice f-strings para implementarlo.




```python
def intercambio(nombre, apellido):
    return f"{apellido}, {nombre}"

print(intercambio("Sergio", "Delgado Quintero"))
print(intercambio("Grace","Hopper"))
print(intercambio("Guido", "Van Rosum"))
```
