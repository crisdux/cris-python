## Convertir HTML a Markdown
Dada un fragmento de HTML que contiene una etiqueta de encabezado , , ...  Se pide transformarlo a su correspondiente versión de Markdown.



```python
def html_a_markdown(html_string):
    corte_1 = html_string[4:]
    corte_2 = corte_1[:-5] ## nodo

    digito_1 = html_string[2:]
    numero = digito_1[:1] ## numero
    
    return f"{int(numero) * '#'} {corte_2}"
   

print(html_a_markdown('<h1>Core</h1>'))
print(html_a_markdown('<h2>Tipos de datos</h2>'))
print(html_a_markdown('<h3>Cadenas de texto</h3>'))
print(html_a_markdown('<h4>hola</h4>'))
```
