## Nombre viceversa
Dado un nombre, obtenga el nombre completo intercambiando nombre y apellidos.




```python
def nombre_viceversa(cadena):
    nombre_index = cadena.find(" ")
    nombre = cadena[:nombre_index]
    apellidos = cadena[nombre_index:]
    return f"{apellidos} {nombre}"

print(nombre_viceversa('John McClane'))
print(nombre_viceversa('Charles Foster Kane'))
print(nombre_viceversa('Terminator T-800'))
```
