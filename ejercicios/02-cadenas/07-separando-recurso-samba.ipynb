## Separando recurso samba
Dada una ruta remota de recurso samba debe separar el nombre del equipo (IP) y la ruta.



```python
def separando_samba(path_samba):
    path_limpio = path_samba[2:]
    host = path_limpio[:path_limpio.find('/')]
    path = path_limpio[path_limpio.find('/'):]
    return host, path

print(separando_samba("//1.1.1.1/aprende/python"))
print(separando_samba("//samba-server/psf/guido"))
print(separando_samba('//8.6.4.2/data/work/upload/'))
```
