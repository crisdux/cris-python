## Quitando primero y último 
Dada una cadena de texto, se pide eliminar el primer y el último caracter de dicha cadena de texto.

-

```python
def primero_ultimo(cadena):
    return cadena[1:][:-1]

print(primero_ultimo("What can I do"))
print(primero_ultimo('Only when I sleep'))
print(primero_ultimo('Runaway'))
print(primero_ultimo('Happiness'))
```
