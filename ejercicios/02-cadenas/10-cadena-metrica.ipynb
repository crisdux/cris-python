## Cadena metrica
Calcule la siguiente métrica para una cadena de texto dada: Número total de caracteres multiplicado por número total de vocales.




```python
def cadena_metrica(cadena):
    vocales = ["a", "e", "i", "o", "u"]
    contador = 0
    for letra in cadena:
        if letra in vocales:
            contador = contador + 1
    return contador * len(cadena)

print(cadena_metrica("ordenador"))
print(cadena_metrica("pantalla"))
print(cadena_metrica("procesador"))
```
