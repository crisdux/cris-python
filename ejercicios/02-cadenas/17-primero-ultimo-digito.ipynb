## Primero el último digito
Determine el primer y el último dígito de una cadena de texto dada.





```python
def primero_ultimo(cadena):
    res = ""
    for caracter in cadena:
        if(caracter.isnumeric()):
            res+=caracter
    return f"{res[0]} {res[-1]}"

print(primero_ultimo('1abc2'))
print(primero_ultimo('pqr3stu8vwx'))
print(primero_ultimo('a1b2c3d4e5f'))
print(primero_ultimo('treb7uchet'))
```
