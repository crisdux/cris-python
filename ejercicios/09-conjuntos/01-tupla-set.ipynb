## Tupla set
Dada una tupla de tuplas (2 valores) obtenga una tupla con dos conjuntos: 

* El primer conjunto con los primeros valores de cada dupla.
*  El segundo conjunto con los segundos valores de cada dupla.




```python
def tupla_set(entrada:tuple):
    primer_set = set()
    segundo_set = set()

    for primero, segundo in entrada:
        primer_set.add(primero)
        segundo_set.add(segundo)
    return primer_set, segundo_set

print(tupla_set( ((4, 3), (8, 2), (7, 5), (8, 2), (9, 1))) )
print(tupla_set(  ((1, 2), ('|', '@'), ('!', '"')) ))
```
