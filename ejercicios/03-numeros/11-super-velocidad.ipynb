 ## Super Velocidad
 El escarabajo tigre es uno de los animales más veloces del planeta llegando a alcanzar el equivalente a 400 km/h (si hablamos de un humano de estatura media).

 Dado que sus distancias son más cortas, convierta la velocidad de km/h a cm/s.

 

```python
def super_rapido(km_h):
    return (km_h*1000*100)/3600

print(f"{super_rapido(1.08):.1f}")
print(f"{super_rapido(4.25):.1f}")
print(f"{super_rapido(0.099):.1f}")



```
