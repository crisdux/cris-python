## El cuadrado rojo
Calcule el Ã¡rea del cuadrado rojo dada la longitud del arco A. Use PI con 2 cifras decimales.
IMPORTANTE: Redondee el resultado a 10 cifras decimales.

![img](https://i.imgur.com/nJrae8n.png)



```python
def cuadrado_rojo(arc_A):
    diametro_circulo = (arc_A*4)/3.14
    radio_circulo = diametro_circulo/2
    area_triangulo = (radio_circulo * radio_circulo)/2
    return area_triangulo * 2

print(f"{cuadrado_rojo(1):.10f}")
print(f"{cuadrado_rojo(0.25):.10f}")
print(f"{cuadrado_rojo(9.99):.10f}")
```
