## Area del anillo
Dado el valor de entrada z calcule el área gris de la siguiente figura.



Notas:
* z es el valor tanto del anillo exterior como del círculo interior.
* Use PI con valor 3.14.
* Obtenga el resultado redondeando a dos cifras decimales.





```python
def area_anillo(z):
    pi = 3.14 
    area_cirulo_pequenio = pi*((z/2)**2)
    area_circulo_grande = pi*((z+(z/2)))**2
    return area_circulo_grande - area_cirulo_pequenio

print(f"{area_anillo(6):.2f}")
print(f"{area_anillo(7):.2f}")
print(f"{area_anillo(8):.2f}")
```
