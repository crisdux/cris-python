## Aproximación del seno
Existe una aproximación al seno de un ángulo x expresado en grados:  



Calcule dicha aproximación utilizando operaciones en Python. Descomponga la expresión en subcálculos almacenados en variables.Tenga en cuenta aquellas expresiones comunes para no repetir cálculos y seguir el principio DRY.                                                                                    ¿Qué tal funciona la aproximación? Compare sus resultados con los "verdaderos":



```python
def sen_aprox(x):
    a = 180 - x 
    A = (4 * x) * a
    B = 40_500 - x * a
    return A/B

print(sen_aprox(90))
print(sen_aprox(45))
print(sen_aprox(50))
    
```
