## Interes compuesto
 Compute el capital final (Cf) a partir del capital inicial (Ci), el tipo de interés (i) y el número de años (n). Fórmula del      │
 interés compuesto:

 

 

```python
def interes_compuesto(amount, rate, years):
    future_amount = amount * (1 + (rate/100))**years
    return future_amount

print(interes_compuesto(10000, 3.5, 7))
print(interes_compuesto(7500, 0.25, 21))
print(interes_compuesto(125000, 0.045, 9))
```
