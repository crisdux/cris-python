## Contando milisegundos
 Un reloj muestra horas, minutos y segundos a partir de medianoche. Calcule el número de milisegundos desde media noche a partir de estos tres parámetros.

  0 <= horas <= 23
  0 <= minutos <= 59
  0 <= segundos <= 59



```python
def contando_miliseguendos(h, m, s):
    horas = h*3_600_000
    minutos = m * 60_000 
    segundos = s * 1_000
    return horas + minutos + segundos

print(contando_miliseguendos(0,1,1))
print(contando_miliseguendos(2,1,0))
print(contando_miliseguendos(10,45,30))
```
