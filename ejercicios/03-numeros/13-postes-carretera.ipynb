## Postes en la carretera
Hay postes cerca de la carretera. La distancia entre los postes es la misma y el ancho de los postes también es el mismo.
Se aceptan tres parámetros de entrada:

* Número de postes.
* Distancia entre postes (en metros) 
* Ancho del poste (centímetros).

Calcule la distancia (en centímetros) entre el primer y el último poste (quitando el ancho del primer y del último poste).








```python
def postes_carretera(num_postes, distancia, ancho_postes):
    total_ancho_postes = num_postes*ancho_postes # cm
    total_espacio_postes = (num_postes - 1) * distancia * 100 
    total_distancia = (total_ancho_postes + total_espacio_postes) - ancho_postes * 2
    return total_distancia

print(postes_carretera(10,5,30))
print(postes_carretera(2, 1.5, 50))
print(postes_carretera(5, 2.25, 75))
```
