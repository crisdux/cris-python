## Distancia euclidiana
Calcule la distancia euclídea entre dos puntos (x1, y1) y (x2, y2).



```python
import math 
def distancia_euclidiana(x1, x2, y1, y2):
    X = (x2 - x1)**2
    Y = (y2 - y1)**2
    return math.sqrt(X+Y)

print(distancia_euclidiana(3,-7,5,-4))
print(distancia_euclidiana(2.1,-2.3,4.3,8.5))
print(distancia_euclidiana(0.1,-0.1,0.2,-0.2))
```
