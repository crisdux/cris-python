## Ecuación Cuadratica
Una ecuación de segundo grado se define como ax^2 + bx + c = 0 y, (en determinados casos) tiene dos soluciones:





```python
import math
def ecuacion_cuadratica(a, b, c):
    x1 = (-(b) + ( math.sqrt((b*b) - 4*a*c)) ) / (2*a)
    x2 = (-(b) - ( (b*b) - 4*a*c)**0.5 ) / (2*a)

    return x1, x2

print(ecuacion_cuadratica(4,-6,2))
print(ecuacion_cuadratica(1,2,-3))
print(ecuacion_cuadratica(-1,-2,8))
```
