## Area de un circulo
 Calcule el área de un círculo a partir de su radio. Tenga en cuenta usar PI con 2 cifras decimales.

 

```python
import math
def area_circulo(r):
    return math.pi * r**2

print(f"{area_circulo(4):.2f}")
print(f"{area_circulo(100):.2f}")
print(f"{area_circulo(39.8):.2f}")
```
