## Impuestos
Calcule el precio de un artículo quitándole el IVA indicado.




```python
def calculo_sin_iva(precio_con_iva, iva):
    return precio_con_iva / (1 + (iva/100))

print(f"{calculo_sin_iva(120,7):.2f}")
print(f"{calculo_sin_iva(50, 4.5):.2f}")
print(f"{calculo_sin_iva(1500, 9.75):.2f}")
```
