## len palabras 
Dada una cadena de texto conteniendo palabras separadas por espacios, obtenga una lista que contenga cada palabra junto a su tamaño 



```python
def len_palabras(cadena):
    lista_cadenas = cadena.split(" ")
    res =[f"{cadena} {len(cadena)}" for cadena in lista_cadenas]
    return res

print(len_palabras('todo se transforma'))
print(len_palabras('prueba'))
print(len_palabras('tiene mucha importancia'))
```
