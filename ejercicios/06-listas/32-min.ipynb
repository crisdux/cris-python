## min
Dada una lista de valores numéricos enteros, obtenga su mínimo valor.
Prohibido utilizar:
* La función "built-in" max().
* La función "built-in" min().
* La función "built-in" sorted().





```python
def min_value(values):
    while values:
        min_value = values[0]
        for value in values:
            if value < min_value:
                min_value = value
        return min_value

print(min_value([-11, 10, -6, 15, -1]))
print(min_value([5, 9, -6, 9, -2, -9, -7, 2]))
print(min_value( [-7, -5, -5, -8, -3]))
```
