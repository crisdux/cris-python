## Todos iguales
 Dada una lista de entrada, devuelva True si todos sus elementos son iguales o False en otro caso.

 

```python
def todos_iguales(items):
    res = []
    for i in items:
        if len(res) == 0 or i == res[-1]:
            res.append(i)
    return len(res) == len(items)
print(todos_iguales([1, 1, 1, 1, 1, 1]))
print(todos_iguales([1, 1, 1, 1, 2]))
print(todos_iguales(['a', 'b', 'c']))
        
```
