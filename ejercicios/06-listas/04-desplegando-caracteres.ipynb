## Desplegando caracteres
Dada una lista de strings, obtenga otra lista que contenga todos los caracteres de cada uno de los strings de la lista de entrada    



```python
def desplegar_caracteres(textos):
    return list("".join(textos))
    

print(desplegar_caracteres(['uno', 'dos', 'tres']))
print(desplegar_caracteres(['', '', '']))
print(desplegar_caracteres(['abc', 'abc', 'abc']))
print(desplegar_caracteres(['XZ', 'VW']))
print(desplegar_caracteres(['alone']))
```
