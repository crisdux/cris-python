## Cuboides
Como datos de entrada tendrá dos listas representando las dimensiones de sendos cuboides:



Debe calcular la diferencia de volumen entre los dos cuboides como valor positivo, indpendientemente de qué cuboide sea mayor. 





```python
def cuboides(cuboide_1, cuboide_2):
    vol_cub_1 = 1
    vol_cub_2 = 1

    for cub_1 in cuboide_1:
        vol_cub_1 *= cub_1

    for cub_2 in cuboide_2:
        vol_cub_2 *= cub_2 
    
    return vol_cub_1 - vol_cub_2 if vol_cub_1 > vol_cub_2 else vol_cub_2 - vol_cub_1

print(cuboides([2,2,3],[5,4,1]))
print(cuboides([3, 4, 2], [9, 1, 6]))
print(cuboides([1, 1, 1],[1,1,1]))
```
