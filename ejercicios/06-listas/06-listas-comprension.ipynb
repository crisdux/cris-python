## Aplica la funcion
Utilizando listas por comprensión, cree una lista que contenga el resultado de aplicar la función f(x) = 3x + 2 para el rango de x dado en los valores de entrada.

Ejemplo: Si xmin=0 y xmax=5, la lista de valores resultante debería ser: [2, 5, 8, 11, 14]



```python
def listas_comprension(xmin, xmax):
    return [numero * 3 + 2 for numero in range(xmin, xmax+1)]

print(listas_comprension(0, 10))
print(listas_comprension(-7, 0))
print(listas_comprension(131, 134))
print(listas_comprension(0, 1))
```
