## Calculadora logica
Dada una lista de valores booleanos y un operador lógico, calcule el resultado booleano tras aplicar el operador a la lista de valores de entrada de forma secuencial.

Tenga en cuenta que los dos únicos posibles operadores de entrada son and y or y que siempre van a aparecer en minúsculas.



```python
def calculadora_logica(values:list, oper:str):
    values_string = [str(value) for value in values]
    exp = ""
    for index, value in enumerate(values_string):
        if index != len(values_string)-1:
            exp += f"{value} {oper} "
        else:
            exp += value
    return eval(exp)

print(calculadora_logica([True, True, False], "and"))
print(calculadora_logica([True, True, False], "or"))
print(calculadora_logica([True, True, True], "and"))
print(calculadora_logica([False, False, False], "and"))
```
