## Suma diagonal
Escriba un programa en Python que acepte una lista de listas representando una matriz numérica de valores enteros y compute la suma de los elementos de la diagonal principal.

* Si no se trata de una Matriz cuadrada devuelva None.
* Puede afrontar el ejercicio en versión clásica o con listas por comprensión. ¡O incluso ambas!



```python
def suma_diagonal(matrix):
    if len(matrix) != len(matrix[0]):
        return None
    suma = 0
    for i in range(len(matrix)):
        suma += matrix[i][i]
    return suma

print(suma_diagonal([[4, 6, 1], [2, 9, 3], [1, 7, 7]]))
print(suma_diagonal( [[1, 1], [1, 1]]))
print(suma_diagonal([[1, 2, 3, 4], [4, 5, 6, 7], [8, 9, 10, 11], [4, 2, 1, 8]]))
print(suma_diagonal([[1, 1, 1], [2, 2, 2], [3, 3, 3], [4, 4, 4]]))
```
