## Suma inversa
Dada una lista de valores num√©ricos, obtenga la suma de los valores invertidos.



```python
def suma_inversa(numbers):
    return sum([number*-1 for number in numbers])

print(suma_inversa([1, 2, 3, 4, 5]))
print(suma_inversa([-1, 2, -3, 4, -5]))
print(suma_inversa([1, -2, 3, -4, 5]))
```
