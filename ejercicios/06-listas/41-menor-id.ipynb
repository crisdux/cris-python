## Id sin usar
Dada una lista de números enteros positivos representando identificadores únicos de objetos, determine cuál es el menor identificador sin usar.
* Puede suponer que no habrán valores repetidos en la lista de entrada.
* Los identificadores "válidos" empiezan en 1.
  


```python
def id_sin_usar(ids):
    ids_sort = sorted(ids)
    ids_aux = [ id_aux for id_aux in range(1, len(ids_sort)+1)]
    if ids_sort == ids_aux:
        return ids_sort[-1] + 1
    else:
        for id, id_aux in zip(ids_sort, ids_aux):
            if id != id_aux:
                return id_aux
                


print(id_sin_usar([3, 1, 8, 4, 9]))
print(id_sin_usar([7, 2, 12, 21, 5]))
print(id_sin_usar([1, 2, 3, 5, 10, 7]))
print(id_sin_usar([4, 3, 2, 1]))
```
