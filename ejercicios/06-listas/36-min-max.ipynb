## min max
Dada una lista de valores numéricos, encuentre el valor máximo y el valor mínimo de la misma. No se pueden utilizar las funciones "built-in" max() y min(), y tampoco sorted().



```python
def min_value(values):
    while values:
        min_value = values[0]
        for value in values:
            if value < min_value:
                min_value = value
        return min_value

def max_value(values):
    while values:
        max_value = values[0]
        for value in values:
            if value > max_value:
                max_value = value
        return max_value

def min_max(values):
    res_min = min_value(values)
    res_max = max_value(values)
    return res_min, res_max

print(min_max([4, 6, 2, 1, 9, 63, -134, 566]))
print(min_max([-52, 56, 30, 29, -54, 0, -110]))
print(min_max([42, 54, 65, 87, 0]))
print(min_max([5]))
```
