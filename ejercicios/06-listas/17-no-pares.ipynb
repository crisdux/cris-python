## No pares 
 Dada una lista, genere otra lista eliminando los elementos que ocupan posiciones pares (tal y como lo vería un "ser humano no programador").

 

```python
def no_pares(items):
    res = []
    for index, item in enumerate(items, start=1):
        if index % 2 != 0:
            res.append(item)
    return res

print(no_pares([1, 2, 1, 2, 1, 2]))
print(no_pares(['a', 'b', 'a', 'b', 'a']))
print(no_pares([1, 1, 2, 2, 3, 3]))
```
