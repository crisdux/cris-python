## Iniciales nombre
Dado un nombre y apellidos en formato "apellidos, nombre", obtenga las iniciales de dicha persona pasadas a mayúsculas y con punto al final

* El nombre puede tener uno o dos elementos.
* El apellido puede tener uno o dos elementos.



```python
def iniciales_nombre(full_name):
    apellidos = full_name.split(",")[0].split(" ")
    nombres = full_name.split(",")[1].strip().split(" ")
    lista_apellidos = []
    lista_nombres = []
    for apellido in apellidos:
        lista_apellidos.append(apellido[0].upper())

    for nombre in nombres:
        lista_nombres.append(nombre[0].upper())
    
    if len(lista_nombres) == 2 and len(lista_apellidos) == 2:
        return f"{lista_nombres[0]}.{lista_apellidos[0]}.{lista_apellidos[1]}."
    elif len(lista_nombres) == 1 and len(lista_apellidos) == 1:
        return f"{lista_nombres[0]}.{lista_apellidos[0]}."
    else:
        return f"{lista_nombres[0]}.{lista_apellidos[0]}.{lista_apellidos[1]}."

print(iniciales_nombre('Delgado Quintero, sergio'))
print(iniciales_nombre('sánchez, María'))
print(iniciales_nombre('Prado López, Ana Belén'))
```
