## Troceado
Partiendo de una cadena de texto con números separados por comas, obtenga una nueva cadena de texto donde se elimine el primer y el último número, y los elementos aparezcan separados por espacios.



```python
def troceado(numbers):
    numbers_list = numbers.split(",")
    if len(numbers_list) <=2:
        return ""
    numbers_list.pop(0)
    numbers_list.pop(-1)
    return " ". join(numbers_list)


print(troceado('1,2,3'))
print(troceado('1,2,3,4'))
print(troceado('1,2,3,4,5'))
print(troceado('1,20,300,40,5'))
print(troceado(''))
print(troceado('1'))
print(troceado('1,2'))
```
