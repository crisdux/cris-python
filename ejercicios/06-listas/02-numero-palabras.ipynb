## Contando palabras
Sabiendo que la longitud de una lista se calcula igual que la longitud de una cadena de texto, obtenga el n√∫mero de palabras que contiene una cadena de texto dada.



```python
def contador_palabras(cadena):
    return len(cadena.split(" "))

print(contador_palabras('Before software can be reusable it first has to be usable'))
print(contador_palabras('You should prefer readability over performance in almost all cases'))
print(contador_palabras('Models are not right or wrong they are more or less useful'))
```
