## palabras inversas 
Dado un string que contiene palabras separadas por espacio, obtenga otro string donde las palabras estén en orden inverso. En la cadena de salida todo el texto debe quedar en minúsculas.



```python
def palabras_inversas(cadena):
    cadena_lower = cadena.lower()
    lista_cadenas = cadena_lower.split(" ")
    return " ".join(list(reversed(lista_cadenas)))

print(palabras_inversas('Hello World'))
print(palabras_inversas('Esto es Python'))
print(palabras_inversas('último caso de comprobación'))
```
