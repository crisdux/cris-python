## min con max
Dada una lista de valores numéricos enteros, obtenga su mínimo valor.

Implementa la solución usando la función "built-in" max().

Prohibido utilizar:
*  La función "built-in" min().
*  La función "built-in" sorted().




```python
def min_max(values):
    copy_values = values.copy()
    while len(copy_values) != 1:
        max_value = max(copy_values)
        copy_values.remove(max_value)
    return copy_values[0]

print(min_max([-11, 10, -6, 15, -1]))
print(min_max([5, 9, -6, 9, -2, -9, -7, 2]))
print(min_max([-7, -5, -5, -8, -3]))
        
```
