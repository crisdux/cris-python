## Consecutivos repetidos
Dada una lista, genere otra lista eliminando los elementos duplicados consecutivos.



```python
def eliminar_repetidos_consecutivos(lista):
    resultado = []
    for i in lista:
        if len(resultado) == 0 or i != resultado[-1]:
            resultado.append(i)
    return resultado

print(eliminar_repetidos_consecutivos([1, 0, 0, 0, 1, 1, 0, 1, 1, 1, 0, 0]))
```
