## Potencias
Dado un valor entero positivo n compruebe que se cumple la siguiente igualdad:



El objetivo es calcular ambos lados de la igualdad. Por ejemplo, para n=3 se cumple:






```python
def potencias(n):
    res_1 = 0
    res_2 = sum([ value ** 3 for value in range(1, n+1)])
    for value in range(1, n+1):
        res_1 += value

    return res_1**2, res_2

print(potencias(1))
print(potencias(2))
print(potencias(3))
print(potencias(4))
print(potencias(5))
```
