## Listas inversas
Dado un número entero no negativo, genere una lista con los dígitos de dicho número en orden inverso.



```python
def listas_inversas(n):
    rev = (list(str(n)))[::-1]
    return [int(n) for n in rev]



print(listas_inversas(35231))
print(listas_inversas(1))
print(listas_inversas(4287690))
```
