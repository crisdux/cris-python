## isogramas
Determine si una cadena de texto dada es un isograma, es decir, no se repite ninguna letra.

No se puede utilizar la funci√≥n count()

Los guiones medios no cuentan como caracter repetido.




```python
def isograma(cadena):
    aux = "" 
    str_to_list = []
    for i in cadena:
        if i != "-":
            str_to_list.append(i)

            
    for index, letra in enumerate(str_to_list):
        aux = str_to_list.pop(index)
        if (aux in str_to_list) and ("-" not in str_to_list):
            return False
        aux = ""
    return True

print(isograma('lumberjacks'))
print(isograma('background'))
print(isograma('downstream'))
print(isograma('six-year-old'))
print(isograma('circus'))
print(isograma('fantastical'))
print(isograma('left-hand-side'))
print(isograma('symmetrical'))
```
