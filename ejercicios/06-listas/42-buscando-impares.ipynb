## Buscando impares
Devuelva una lista únicamente con los números impares que vienen en la entrada. Tenga en cuenta que pueden venir valores repetidos en la entrada.



```python
def buscando_impares(values):
    res = []
    for value in values:
        if value not in res:
            res.append(value)
    return [i for i in res if i % 2 != 0]

print(buscando_impares([1, 2, 3, 4, 5]))
print(buscando_impares([1, 1, 2, 2, 3, 3]))
print(buscando_impares([2, 4, 6]))
```
