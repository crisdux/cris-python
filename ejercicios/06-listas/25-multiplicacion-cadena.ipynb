## Cadena multiplicacion
Dada una lista de valores numéricos enteros, obtenga el resultado de multiplicar todos los valores en orden.





```python
def multiplicacion_cadena(numeros):
    res = 1
    for numero in numeros:
        res *= numero
    return res

print(multiplicacion_cadena([1, 2, 3, 4]))
print(multiplicacion_cadena([101, 102, 103, 104]))
print(multiplicacion_cadena([]))
print(multiplicacion_cadena([1]))
```
