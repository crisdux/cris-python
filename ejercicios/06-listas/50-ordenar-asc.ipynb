## ordenar
Ordene la lista de números dada de manera ascendente sin utilizar la función sorted().

IMPORTANTE: No modifique la lista de entrada. En vez de eso haga una copia en otra lista antes de modificar.




```python
def ordenar_asc(numbers):
    copy_numbers = numbers.copy()
    res = []
    for _ in range(len(copy_numbers)):
        min_number = min(copy_numbers)
        res.append(min_number)
        copy_numbers.remove(min_number)
    return res

print(ordenar_asc([4, 2, 9, 7, 1, 8]))
print(ordenar_asc([21, 11, 5, 34, 42, 7, 16, 3, 51, 18]))
print(ordenar_asc([3, 2, 2, 5, 7, 1, 7, 9, 2, 9, 9, 4]))
print(ordenar_asc([0]))
print(ordenar_asc([1, 0]))
```
