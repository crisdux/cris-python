## Año biciesto
Dada una variable year con un valor entero, compruebe si dicho año es bisiesto o no lo es.
Un año es bisiesto en el calendario Gregoriano, si es divisible entre 4 y no divisible entre 100, o bien si es divisible entre 400. Puede hacer la comprobación en esta lista de años bisiestos.



```python
def biciesto(year):
    if (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

print(biciesto(1900))
print(biciesto(1904))
print(biciesto(1983))
print(biciesto(1992))
print(biciesto(2000))
print(biciesto(2002))
print(biciesto(2052))
print(biciesto(2100))
```
