## Producto Cartesiano 
A partir de dos cadenas de texto compute el producto cartesiano letra a letra, dando como resultado un nuevo string completo.



```python
def producto_cartesiano(cadena_1, cadena_2):
    res = ""
    for i in cadena_1:
        for j in cadena_2:
            res += f"{i}{j}"
    return res 

print(producto_cartesiano("xy", "$#"))
print(producto_cartesiano("abc", "123"))
print(producto_cartesiano("ho", "ho"))
print(producto_cartesiano("ðŸ˜€ðŸ˜¡", "ðŸ¤”ðŸ˜”"))
```
