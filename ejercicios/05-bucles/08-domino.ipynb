## Domino
Escriba un programa que muestre por pantalla todas las fichas del domin. La ficha
en blanco se puede representar con un 0 (solucin).




```python
def domino():
    for i in range(0,7):
        print("")
        for j in range(7):
            if i <= j:
                print(f"{i}|{j}", end =" ")

print(domino())
```
