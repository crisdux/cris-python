## Ascii
Escriba un programa que muestre (por filas) la Tabla ASCII, empezando con el código
33 y terminando con el 127 (solución)



```python
def codigo_ascci():
    contador = -1
    for i in range(33,128):
        contador = contador + 1
        if contador % 5 == 0:
            print("\n")
        print(f"{i:03} {chr(i)}", end="\t")
        
            

print(codigo_ascci())
```
