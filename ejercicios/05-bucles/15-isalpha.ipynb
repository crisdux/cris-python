## isaplha
Escriba un programa en Python que acepte una cadena de texto e indique si todos sus caracteres son alfabéticos.

No usar la función isalpha()

Usar una constante: ALPHABET = 'abcdefghijklmnopqrstuvwxyz'



```python
def fn_isalpha(cadena):
    ALPHABET = 'abcdefghijklmnopqrstuvwxyz'
    res = []
    for i in cadena.lower():
        if i not in ALPHABET:
            res.append(i)
    return True if len(res) == 0 else False

print(fn_isalpha('hello-world'))
print(fn_isalpha('Computer'))
print(fn_isalpha('first_in_first_out'))
print(fn_isalpha('Development'))
print(fn_isalpha('Programming is fun!'))
```
