## Hamming
Calcule la distancia hamming entre dos cadenas de texto dadas.



```python
def hamming(cadena_1, cadena_2):
    if len(cadena_1) != len(cadena_2):
        return -1

    list_1 = list(cadena_1)
    list_2 = list(cadena_2)
    
    contador = 0
    for i in range(len(list_1)):
        if list_1[i] != list_2[i]:
            contador += 1
    return contador

print(hamming("0001010011101","0000110010001"))
print(hamming("abc", "abcd"))
print(hamming("teclado", "techado"))
print(hamming("esperanza", "esmeralda"))
```
