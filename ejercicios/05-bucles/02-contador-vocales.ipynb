## Contador de vocales
Dada una cadena de texto, indique el número de vocales que tiene.
No se puede utilizar la función count()



```python
def contador_vocales(cadena):
    cadena_limpia = cadena.lower()
    vocales = "aeiouáéíóú"
    contador_vocales = 0
    for letra in cadena_limpia:
        if letra in vocales:
            contador_vocales += 1
    return contador_vocales

print(contador_vocales('El camión chocó contra el árbol'))
print(contador_vocales('WELCOME HOME'))
print(contador_vocales('Órbita Laica'))
print(contador_vocales('Programar va de pensar, no de escribir'))
print(contador_vocales('Simple es mejor que complejo'))
print(contador_vocales(''))
```
