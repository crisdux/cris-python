## Igualdad potencias
 Dado un valor entero positivo n compruebe que se cumple la siguiente igualdad:



→ El objetivo es calcular ambos lados de la igualdad. Por ejemplo, para n=3 se cumple:






```python
def igualdad_potencias(n):
    suma_izq = 0
    suma_der = 0
    for i in range(1, n+1):
        suma_izq += i
        suma_der += i**3
    
    return suma_izq ** 2, suma_der

print(igualdad_potencias(1))
print(igualdad_potencias(2))
print(igualdad_potencias(3))
print(igualdad_potencias(4))
print(igualdad_potencias(5))
```
