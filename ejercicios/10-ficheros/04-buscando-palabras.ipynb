Dado un fichero de texto y una palabra objetivo se pide indicar el nmero de lnea y el nmero de columna en la que se encuentran todas las ocurrencias de dicha palabra en el fichero de entrada. 

La salida del programa ser una lista de tuplas. Cada tupla contiene dos valores: el nmero de lnea y el nmero de columna. 

Se puede suponer que las palabras se separan por un nico espacio, pero hay que tener en cuenta la posible existencia de los siguientes caracteres como frontera de palabras: .,:;()'!-  

La bsqueda debe funcionar aunque no coincidan maysculas y minsculas.



```python
def encontrar_palabras(fichero, palabra_objetivo):
    palabra_obejtivo = palabra_objetivo.lower()
    ocurrencias = []
    delimitadores = " .,;:()'!-"

    with open(fichero, "r", encoding="utf-8") as archivo:
        for num_linea, linea in enumerate(archivo, start=1):
            linea_lower = linea.lower()
            columna = 1
            palabra_actual = ""

            for i, caracter in enumerate(linea_lower): #no se acaba el mundo
                if caracter in delimitadores or caracter == "\n":
                    if palabra_actual == palabra_objetivo:
                        ocurrencias.append((num_linea, columna))
                    columna = i+ 2
                    palabra_actual = ""
                else:
                    if palabra_actual == "":
                        columna = i + 1
                    palabra_actual += caracter
            if palabra_actual == palabra_objetivo:
                ocurrencias.append((num_linea, columna))
    return ocurrencias

    return ocurrencias

print(encontrar_palabras("data/cancion.dat", "mundo"))
```
