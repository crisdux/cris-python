## Valores iguales
Acepte un diccionario como entrada y compruebe si todos sus valores son iguales o no.



```python
def valores_iguales(items:dict):
    if len(items) < 1:
        return True
    values = list(items.values())
    primer_valor = values[0]
    res = None
    for i in values:
        if i != primer_valor:
            res = False
            break
        res = True
    return res

print(valores_iguales({'a': 1, 'b': 1, 'c': 1, 'd': 1}))
print(valores_iguales({'a': 1, 'b': 2, 'c': 3, 'd': 4}))
print(valores_iguales({1: 'a', 2: 'b', 3: 'c', 4: 'd'}))
print(valores_iguales({1: 'a', 2: 'a', 3: 'a', 4: 'a'}))
print(valores_iguales({}))
```
