## Concatenar diccionarios
Dados dos diccionarios de entrada, escriba un programa en Python que mezcle ambos diccionarios en uno único de salida, sin usar métodos de mezcla definidos en Python como:
* {**a, **b}
* a | b
* a.update(b)



```python
def concatenar_dict(d1:dict, d2:dict):
    copia_d1 = d1.copy()
    for clave, valor in d2.items():
        copia_d1[clave] = valor
    return copia_d1


print(concatenar_dict({'a': 1, 'b': 2}, {'a': 3, 'c': 4}))
print(concatenar_dict({0: 1, 1: 0}, {0: 0, 1: 1}))
print(concatenar_dict({}, {'x': 2.1, 'y': 3.4}))
print(concatenar_dict({}, {}))
```
