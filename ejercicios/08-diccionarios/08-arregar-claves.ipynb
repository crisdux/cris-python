## arreglar claves
A partir de un diccionario, obtenga un nuevo diccionario eliminando los espacios de sus claves respetando los valores  correspondientes.



```python
def arreglar_claves(items):
    res = {}
    clave_limpia = ""
    for clave, valor in items.items():
        clave_limpia = ""
        for caracter in clave:
            if caracter != " ":
                clave_limpia += caracter
        res[clave_limpia] = valor
    return res

print(arreglar_claves({'S  001': ['Math', 'Science'], 'S    002': ['Math', 'English']}))
```
