## Contador letras
Usando un diccionario, cuente el número de veces que se repite cada letra en una cadena de texto dada. 

NOTA: No se puede usar la función "count".



```python
def contador_letras(cadena):
    res = {}
    for letra in cadena:
        if letra not in res:
            res[letra] = 1
        else:
            res[letra] = res[letra] + 1
    return res

print(contador_letras('boom'))
print(contador_letras('ada'))
print(contador_letras('tree'))
print(contador_letras('debugging'))
print(contador_letras(''))
```
