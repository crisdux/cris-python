## Regresa cambio 
 Un cliente compra un artículo en una tienda con dinero suficiente (mayor o igual) que el importe del artículo. Por lo tanto hay que devolver el cambio. La tienda dispone de una serie concreta de monedas/billetes. El objetivo del ejercicio es devolver e cambio al cliente empezando por la moneda/billete más grande y llegando hasta la más pequeña.
 * Si el dinero es justo hay que devolver un diccionario vacío.
 * Si el cambio no es posible hay que devolver None



```python
def regresa_cambio(cambio:float, caja:list):
    caja_ordenada = sorted(caja, reverse=True)
    if cambio == 0:
        return {}
    if cambio < caja_ordenada[0]:
        return None
    cambio_sucio = {}
    for moneda in caja_ordenada:
        cociente = cambio // moneda
        cambio = cambio % moneda
        cambio_sucio[moneda] = cociente
    
    cambio_limpio = {}
    for clave, valor in cambio_sucio.items(): ## {5: 4, 2: 0, 1: 0}
        if valor == 0:
            continue
        cambio_limpio[clave] = valor
    return cambio_limpio


print(regresa_cambio(20, [5, 2, 1]))
print(regresa_cambio(7, [2, 1, 0.5]))
print(regresa_cambio(13.5, [5, 2, 0.5]))
print(regresa_cambio(11, [0.1, 0.5, 2]))
print(regresa_cambio(0, [0.5, 0.2, 0.1]))
print(regresa_cambio(4.75, [1, 5, 0.2]))
```
