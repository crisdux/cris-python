## Tuplas por entrada



```python
def tupla_input():
    x = tuple(input("Escribe la primera coordenada"))
    y = tuple(input("Escribe la segunda coordenada"))
    ubicacion = tuple(input("Escribe una ubicacion"))
    res = x + y + ubicacion
    mi_tupla = tuple(res)
    return mi_tupla

print(tupla_input())
```
