## Face Emoji
Dada una cadena de texto que represente una carita emoji se pide que se devuelva el emoji en cuestión.



```python
def face_emoji(emoji_text):
    match emoji_text.lower():
        case "happy":
            return "😀"
        case "sad":
            return "😔"
        case "angry":
            return "😡"
        case "pensive":
            return "🤔"
        case "surprised":
            return "😧"

print(face_emoji("sad"))
print(face_emoji("HappY"))
```
