## El gran libro
Complete la función para que encuentre el promedio de las tres puntuaciones que se le pasaron y devuelva el valor de la letra asociada con esa calificación.



Los valores estimados están todos entre 0 y 100. No es necesario comprobar si hay valores negativos o valores superiores a 100.

```python
def gran_libro(valor):
    if valor >= 90 and valor <= 100:
        return "A"
    elif valor >= 80 and valor < 90:
        return "B"
    elif valor >= 70 and valor < 80:
        return "C"
    elif valor >= 60 and valor < 70:
        return "D"
    else:
        return "F"

print(gran_libro(99))
print(gran_libro(85))
print(gran_libro(74))
print(gran_libro(63))
print(gran_libro(55))
```
