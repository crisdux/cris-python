## Atajos
Combinando las teclas CTRL y ALT podemos conseguir muchos shortcuts (atajos de teclado) en sistemas tipo Linux.
Se pide identificar las combinaciones de teclas indicando la función que realizan.





```python
def atajos(key_1="", key_2="", key_3=""):
    key_1_clean = key_1.upper()
    key_2_clean = key_2.upper()
    key_3_clean = key_3.upper()

    if (key_1_clean == "CTRL" and key_2_clean == "ALT" and key_3_clean=="T"):
        return "open terminal"
    elif (key_1_clean == "CTRL" and key_2_clean == "ALT" and key_3_clean=="L"):
        return "Look Screen"
    elif (key_1_clean == "CTRL" and key_2_clean == "ALT" and key_3_clean=="D"):
        return "show desktop"
    elif (key_1_clean == "ALT" and key_2_clean == "F2"):
        return "Look Screen"
    elif (key_1_clean == "CTRL" and key_2_clean == "Q"):
        return "Close Window"
    elif (key_1_clean == "CTRL" and key_2_clean == "ALT" and key_3_clean == "DEL"):
        return "Log out"
    

print(atajos("CTRL", "ALT", "T"))
print(atajos("CTRL", "ALT", "L"))
print(atajos("CTRL", "ALT", "D"))
print(atajos("ALT", "F2"))
print(atajos("CTRL", "Q"))
print(atajos("CTRL", "ALT", "DEL"))
    
```
