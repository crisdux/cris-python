## Calificaciones 
Cree una función que tome un número como argumento y devuelva una calificación basada en ese número.



```py
grader(0)   should be "F"
grader(1.1) should be "F"
grader(0.9) should be "A"
grader(0.8) should be "B"
grader(0.7) should be "C"
grader(0.6) should be "D"
```


````python
def calificaciones(calificacion):
    if calificacion >1 or calificacion <0.6:
        return "F"
    elif calificacion >= 0.9:
        return "A"
    elif calificacion >= 0.8:
        return "B"
    elif calificacion >= 0.7:
        return "C"
    else:
        return "D"

print(calificaciones(0))
print(calificaciones(1.1))
print(calificaciones(0.9))
print(calificaciones(0.8))
print(calificaciones(0.7))
print(calificaciones(0.6))
````
